```python
from sklearn.model_selection import train_test_split
from sklearn.compose import ColumnTransformer
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import OneHotEncoder
from sklearn.preprocessing import MinMaxScaler
from sklearn.pipeline import Pipeline,make_pipeline
from sklearn.feature_selection import SelectKBest,chi2
from sklearn.tree import DecisionTreeClassifier
```

```python
df = pd.read_csv('train.csv')
```

```python
df.drop(columns=['PassengerId','Name','Ticket','Cabin'],inplace=True)
```

```python
X_train,X_test,y_train,y_test = train_test_split(df.drop(columns=['Survived']),
                                                 df['Survived'],
                                                 test_size=0.2,
                                                random_state=42)
```

```python
# WHere the [2] & [6] Are index num of columns
# we want to carry reaming columns with it so we are using these , to passthrpuh the columns so

trf1 = ColumnTransformer([
    ('impute_age',SimpleImputer(),[2]),
    ('impute_embarked',SimpleImputer(strategy='most_frequent'),[6])
],remainder='passthrough')
```

```python
trf2 = ColumnTransformer([
    ('ohe_sex_embarked',OneHotEncoder(sparse=False,handle_unknown='ignore'),[1,6])
],remainder='passthrough')
```

```python
# Scaling # After Adding Columns we get all the table numbers so we do these
trf3 = ColumnTransformer([
    ('scale',MinMaxScaler(),slice(0,10))
])
```

```python
trf4 = SelectKBest(score_func=chi2,k=8)
```

```python
# train the model
trf5 = DecisionTreeClassifier()
```

```python
pipe = Pipeline([
    ('trf1',trf1),
    ('trf2',trf2),
    ('trf3',trf3),
    ('trf4',trf4),
    ('trf5',trf5)
])
```

```python
pipe.fit(X_train,y_train)
```

```python
pipe.named_steps
```

```python
# Display Pipeline

from sklearn import set_config
set_config(display='diagram')
```

```python
# Predict
y_pred = pipe.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score
accuracy_score(y_test,y_pred)
```
